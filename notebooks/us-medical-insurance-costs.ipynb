# U.S. Medical Insurance Costs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

notebook_path = Path.cwd()
data_file_path = notebook_path.parent / 'data' / 'insurance.csv'

# 1. Daten laden
df = pd.read_csv(data_file_path)
# 2. Analyse / visualisieren
regional_avg = df.groupby('region')['charges'].mean().reset_index()

# 3. Plot erstellen
plt.figure(figsize=(8, 5))
sns.barplot(x='region', y='charges', data=regional_avg, palette='viridis')
plt.title('Durchschnittliche Krankenversicherungskosten pro Region')
plt.xlabel('Region')
plt.ylabel('Durchschnittliche Kosten ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
